```python
#se importan las librerias a usar
import csv
import json
import numpy as np
import pandas as pd
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#se define una funcion para leer formatos json
def leer_contenido(ruta_archivo):
    with open(ruta_archivo, 'r', encoding='utf-8') as archivo:
        contenido = json.load(archivo)
    return contenido
```

```python
#se lee el contenido del formato json
datos = leer_contenido("data/transformaciones.json")
```

```python
#se realizan las primeras impresiones de los datos en el dataframe
data_frame = pd.DataFrame(datos)
data_frame.head()
```
